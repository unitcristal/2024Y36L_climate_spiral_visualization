```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data setup (a small part of your dataset)
data = {
    'Year': [1880, 1880, 1880, 1880, 1880, 1881, 1881, 1881],
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jan', 'Feb', 'Mar'],
    'Anomaly': [-0.2, -0.25, -0.09, -0.16, -0.09, -0.19, -0.15, 0.02]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Combining Year and Month to create a time axis
df['Date'] = pd.to_datetime(df['Year'].astype(str) + df['Month'], format='%Y%b')

# Plotting
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Anomaly'], marker='o', linestyle='-', color='b')
plt.title('Temperature Anomalies Over Time')
plt.xlabel('Date')
plt.ylabel('Anomaly')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 날짜 데이터를 datetime 형식으로 변환 (만약 이미 datetime이 아닌 경우)
df['Date'] = pd.to_datetime(df['Date'])

# 그래프 그리기
plt.plot(df['Date'], df['Anomaly'], marker='o', linestyle='-', color='blue')

# X축에 날짜 형식 맞추기
plt.gcf().autofmt_xdate()

# 제목과 라벨
plt.title("Temperature Anomalies Over Time")
plt.xlabel("Date")
plt.ylabel("Anomaly")

# X축 눈금 조정 (회전 각도 조정)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
